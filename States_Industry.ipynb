```
df1.reset_index()
df1.set_index(['State','Key_word'],inplace=True)
```

```
index=df1.index
name=str(index[0][0])
```

```
path='/content/drive/Shared drives/Build with AI/Final Plots'
try:
   os.mkdir(path)
except OSError:
  print ("Creation of the directory %s failed" % path)
else:
  print ("Successfully created the directory %s " % path)
```

```
for x in index:
  if x[0]==name:
    plt.plot(df1.loc[x,:],label=x[1])
    plt.title(name)
    plt.xticks(rotation=90)
    plt.xlabel('Month')
    plt.ylabel('Unemployment Rate')
  else:
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5),ncol=2, fancybox=True)
    plt.tight_layout()
    d1=os.path.join(path,name)
    plt.savefig(d1+'.png')
    plt.clf()
    name=x[0]

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5),ncol=2, fancybox=True)
plt.tight_layout()    
d1=os.path.join(path,name)
plt.savefig(d1+'.png') 
plt.clf()
```

```
df2=pd.read_csv('/content/drive/Shared drives/Build with AI/state_industryclean.csv')
```

```
sectors=df2['Key_word']
sectors=sectors[sectors.duplicated(keep=False)]
sectors=sectors.drop_duplicates()
```

```
sectors
```

```
df2.reset_index()
df2.set_index(['Key_word','State'],inplace=True)
df2 = df2.apply(pd.to_numeric, errors='coerce')
```

```
index=df2.index
```

```
path='/content/drive/Shared drives/Build with AI/Final Plots on sector based'
try:
   os.mkdir(path)
except OSError:
  print ("Creation of the directory %s failed" % path)
else:
  print ("Successfully created the directory %s " % path)
```

```
for j in sectors:
  for x in index:
    if x[0]==j:
      plt.plot(df2.loc[x,:],label=x[1])
  plt.title(str(j))
  plt.xticks(rotation=90)
  plt.xlabel('Month')
  plt.ylabel('Unemployment Rate')
  plt.legend( loc='center left', bbox_to_anchor=(1, 0.5),ncol=3, fancybox=True)
  plt.tight_layout()
  name=str(j)
  d2=os.path.join(path,name)
  plt.savefig(d2+'.png')
  plt.clf()

    
  
```
